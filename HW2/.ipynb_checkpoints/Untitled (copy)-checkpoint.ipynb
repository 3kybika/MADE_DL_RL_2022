Для начала создадим среду:

```python
class TicTacToe(gym.Env):
    def __init__(self, rows_num=ROWS_NUM, cols_num=COLS_NUM, win_len=WIN_LEN):
        self.rows_num = rows_num
        self.cols_num = cols_num
        self.win_len = win_len

        self.reset()

    @staticmethod
    def clone(copy_from):
        new_env = TicTacToe(
            rows_num=copy_from.rows_num, 
            cols_num=copy_from.cols_num,
            win_len=copy_from.win_len
        )

        new_env.board = np.array(copy_from.board, copy=True) 
        new_env.empty_spaces = np.array(copy_from.empty_spaces, copy=True) 
        new_env.board_hash = copy.copy(copy_from.board_hash)
        new_env.game_ended = copy_from.game_ended
        new_env.cur_turn = copy_from.cur_turn

        return new_env

    def get_empty_cells(self, force_update=False):
        if self.empty_spaces is None or force_update:
            res = np.where(self.board == 0)
            self.empty_spaces = np.array(list(zip(res[0], res[1])))
        return self.empty_spaces

    def get_hash(self, force_update=False):
        if self.board_hash is None or force_update:
            self.board_hash = ''.join([
                str(cell + 1) for cell in self.board.reshape(self.rows_num * self.cols_num)
            ])

        return self.board_hash

    def print_board(self):
        print('╭', ('───┬' * self.cols_num)[:-1], '╮', sep='')
        for i in range(0, self.rows_num):
            if i != 0:
                print('├' + ('───┼' * self.cols_num)[:-1] + '┤')
            out = '│ '
            for j in range(0, self.cols_num):
                if self.board[i, j] == 1:
                    out += '✕'
                if self.board[i, j] == -1:
                    out += '◯'
                if self.board[i, j] == 0:
                    out += ' '
                out += ' │ '
            print(out, sep='')
        print('╰' + ('───┴' * self.cols_num)[:-1] + '╯')

    def update(self):
        self.empty_spaces = None
        self.board_hash = None
        self.get_hash()
        self.get_empty_cells()

    def player_turn(self, player, i, j):
        self.board[i, j] = player
        self.update()

    def get_state(self):
        return (self.get_hash(), self.get_empty_cells(), self.cur_turn)

    def int_to_action(self, action_int):
        return (int(action_int / self.cols_num), int(action_int % self.cols_num))

    def action_to_int(self, action):
        return action[0] * self.cols_num + action[1]

    def is_terminal(self):
        cur_marks = np.where(self.board == self.cur_turn)
        cur_position = self.cur_turn

        for i, j in zip(cur_marks[0], cur_marks[1]):
            game_ended = False

            if (
                i <= self.rows_num - self.win_len and
                np.all(self.board[i:i + self.win_len, j] == cur_position)
            ):
                game_ended = True

            if (
                not game_ended and 
                j <= self.cols_num - self.win_len and
                np.all(self.board[i, j:j + self.win_len] == cur_position)
            ):
                game_ended = True

            if (
                not game_ended and
                i <= self.rows_num - self.win_len and j <= self.cols_num - self.win_len and
                np.all(np.array(
                    [self.board[i + k, j + k] == cur_position for k in range(self.win_len)]
                ))
            ):
                game_ended = True

            if (
                not game_ended and
                i <= self.rows_num - self.win_len and j >= self.win_len-1 and
                np.all(np.array([
                    self.board[i + k, j - k] == cur_position for k in range(self.win_len) ]
                ))
            ):
                game_ended = True

            if game_ended:
                self.game_ended = True
                return self.cur_turn

        if len(self.get_empty_cells()) == 0:
            self.game_ended = True
            return 0

        self.game_ended = False
        return None

    def step(self, action):
        if self.board[action[0], action[1]] != 0:
            return self.get_state(), -10, True, {}

        self.player_turn(self.cur_turn, *action)

        reward = self.is_terminal()
        self.cur_turn = -self.cur_turn
        return self.get_state(), 0 if reward is None else reward, reward is not None, {}

    def reset(self):
        self.board = np.zeros((self.rows_num, self.cols_num), dtype=int)
        self.empty_spaces = None

        self.board_hash = None
        self.game_ended = False

        self.cur_turn = 1
```

Изначально я обучал стратегии только на игре друг с другом, так чтобы они выбирали наиболее удачный для себя ход с вероятностью 1 - e. Однако, стратегии ноликов от этого было грустно - крестики побеждали её с крайне большой частовой( Поэтому стратегии сначала обучались с рандомом каждая (чтобы нолики набирались опыта), а затем - друг против друга.

```python
def learn_episode(env, cross_strategy, circle_strategy, alpha=0.009, epsilon=0.1, gamma=1.0):

    env.reset()
    
    cross_prev_action = None
    cross_prev_board_state = None

    circle_prev_action = None
    circle_prev_board_state = None

    while not env.game_ended:
        
        # get current strategy
        if env.cur_turn == 1:
            cur_agent_strategy = cross_strategy
            next_agent_strategy = circle_strategy

            next_agent_prev_action = circle_prev_action
            next_agent_prev_board_state = circle_prev_board_state
        else:
            cur_agent_strategy = circle_strategy
            next_agent_strategy = cross_strategy

            next_agent_prev_action = cross_prev_action
            next_agent_prev_board_state = cross_prev_board_state

        state_hash, empty_cells, cur_turn = env.get_state()
        
        # get current action
        if cur_agent_strategy is None :
            idx = np.random.randint(len(empty_cells))
        else: 
            if state_hash in cur_agent_strategy.Q and np.random.random() > epsilon: 
                # action by strategy
                idx = np.argmax(cur_agent_strategy.Q[state_hash])
            else:
                # action by exploration
                idx = np.random.randint(len(empty_cells))
                if state_hash not in cur_agent_strategy.Q:
                    # explore new state
                    cur_agent_strategy.Q[state_hash] = np.zeros(len(empty_cells))
        if DEBUG:       
            print("cur_agent_strategy: {}".format(cur_agent_strategy.name))
            print("idx: {}".format(idx))
            print("empty_cells: {}".format(empty_cells))
            print("empty_cells[idx]: {}".format(idx))

        cur_action = empty_cells[idx]
        (next_state_hash, _, _), reward, is_ended, _ = env.step(cur_action)

        # train next player straregy
        if next_agent_strategy is not None and next_agent_prev_board_state is not None: 

            gamma_term = 0
            if next_state_hash in next_agent_strategy.Q:
                gamma_term = gamma * np.max(next_agent_strategy.Q[next_state_hash])
            
            if DEBUG:
                print(
                    "next_agent_prev_action: {}".format(
                        next_agent_prev_action
                    )
                )
                print(
                    "next_agent_prev_board_state: {}".format(
                        next_agent_prev_board_state
                    )
                )
                print(
                    "next_agent_strategy.Q[next_agent_prev_board_state]: {}".format(
                        next_agent_strategy.Q[next_agent_prev_board_state]
                    )
                )
                print(
                    "next_agent_strategy.Q[next_agent_prev_board_state][next_agent_prev_action]: {}".format(
                        next_agent_strategy.Q[next_agent_prev_board_state][next_agent_prev_action]
                    )
                )
            prev_q_value = next_agent_strategy.Q[next_agent_prev_board_state][next_agent_prev_action]

            #cur_reward - reward of current agent: cur_turn * reward 
            # if it's crosses turn: 1 * reward, else - (-1) * (-reward)
            cur_reward = cur_turn * reward
            new_q = (cur_reward + gamma_term)
            err = (new_q - prev_q_value)
            next_agent_strategy.Q[next_agent_prev_board_state][next_agent_prev_action] = \
                prev_q_value + alpha * err

        # this is good state for curent agent, because it has won
        if cur_agent_strategy is not None and abs(reward) == 1:
              cur_agent_strategy.Q[state_hash][idx] = abs(reward)

        if env.cur_turn == 1:
            circle_prev_board_state = state_hash
            circle_prev_action = idx
        else:
            cross_prev_board_state = state_hash
            cross_prev_action = idx
```

```python
def play_game(env, cross_strategy, circle_strategy):
    env.reset()

    reward = 0
    is_ended = False

    while not is_ended:
        state_hash, empty_cells, cur_turn = env.get_state()
        # get current strategy
        if cur_turn == 1:
            cur_agent_strategy = cross_strategy
        else:
            cur_agent_strategy = circle_strategy

        # get current action
        if cur_agent_strategy is not None and state_hash in cur_agent_strategy.Q:
            action_idx = np.argmax(cur_agent_strategy.Q[state_hash])
        else:
            action_idx = np.random.randint(len(empty_cells))
        
        # step
        cur_action = empty_cells[action_idx]
        _, reward, is_ended, _ = env.step(cur_action)
        

        if DEBUG:
            print(cur_action)
            env.print_board()
            print(is_ended)
    
    return reward
```

```python
def estimate_reward(env, cross_strategy, circle_strategy, num_episodes=1000):
    acc_reward = 0
    for _ in range(num_episodes):
        acc_reward += play_game(env, cross_strategy, circle_strategy)

    return acc_reward / num_episodes
```

```python
def plot_stats(crosse_scores, circle_scores, test_episodes):
    display.clear_output(wait=True)
    
    fig = plt.figure(figsize=(18, 8))
    plt.xlabel('Number of episodes')
    plt.ylabel('Average reward')
    plt.plot(range(0, len(crosse_scores) * test_episodes, test_episodes), crosse_scores, label='Crosses')
    plt.plot(range(0, len(circle_scores) * test_episodes, test_episodes), circle_scores, label='Circles')
    plt.legend(loc="upper left")
    plt.show()
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
class ReplayMemory():
    def __init__(self, capacity, device):
        self.capacity = capacity
        self.memory = []
        self.position = 0
        self.device = device

    def store(self, exptuple):
        #print(exptuple[1])
        if len(self.memory) < self.capacity:
            self.memory.append(None)
        self.memory[self.position] = exptuple
        self.position = (self.position + 1) % self.capacity
       
    def sample(self, batch_size):
        transitions = random.sample(self.memory, batch_size)
        batch_state, batch_action, batch_next_state, batch_reward = zip(*transitions)
        
        batch_state = Variable(torch.stack(batch_state).to(self.device))
        batch_action = Variable(torch.tensor(batch_action).unsqueeze(1).to(self.device))
        batch_next_state = Variable(torch.stack(batch_next_state).to(self.device))
        batch_reward = Variable(torch.cat(batch_reward).to(self.device))
        
        return batch_state, batch_action, batch_reward, batch_next_state

    def __len__(self):
        return len(self.memory)
```

```python
class AgentDQN:
    def __init__(
        self,
        model,
        n_rows=3,
        epsilon=0.9,
        gamma=0.8,
        buffer_size=1024,
        batch_size=512,
        learning_rate=1e-4,
        device='cpu',
        name=''
    ):
        self.name = name
        self.n_rows = n_rows 
        self.device = device

        self.epsilon = epsilon
        self.gamma = gamma
        self.batch_size = batch_size
        
        self.model = model().to(device)
        self.memory =  ReplayMemory(buffer_size, device)
        self.optimizer = optim.Adam(self.model.parameters(), lr=learning_rate, weight_decay=0.001)
                   
    def learn_on_history(self):
        if len(self.memory) < self.batch_size:
            return
        batch_state, batch_action, batch_reward, batch_next_state = self.memory.sample(self.batch_size)

        Q = (
            self.model(batch_state)
                .gather(1, batch_action)
                .reshape([self.batch_size])
        )
        
        Qmax = self.model(batch_next_state).detach().max(1)[0]
        Qnext = batch_reward + (self.gamma * Qmax)
        
        loss = F.smooth_l1_loss(Q, Qnext)

        self.optimizer.zero_grad()
        loss.backward()
        self.optimizer.step()
        
    def get_action(self, state, strict=False):
        state = state.to(self.device)
        if strict or np.random.random() > self.epsilon:
            return int(self.model(state.unsqueeze(0)).data.max(1)[1])
        else:
            return random.randrange(self.n_rows * self.n_rows)
```

```python
def state_to_tensor(state, cols_num):
    state = np.array([int(x) for x in state])
    
    s_1 = np.where(state > 1, 1, 0).reshape(cols_num, cols_num)
    s_2 = np.where(state < 1, 1, 0).reshape(cols_num, cols_num)
    s_3 = np.where(state == 1, 1, 0).reshape(cols_num, cols_num)
    
    return torch.Tensor(np.stack([s_1, s_2, s_3])).reshape(3, cols_num, cols_num)
```

```python
NOT_IN_EMPTY_REWARD = -10

def learn_episode(env, cross_strategy, circle_strategy, alpha=0.009, epsilon=0.1, gamma=1.0):

    env.reset()
    
    cross_prev_action = None
    cross_prev_state = None

    circle_prev_action = None
    circle_prev_state = None

    while not env.game_ended:
        
        if env.cur_turn == 1:
            cur_agent_strategy = cross_strategy
            cur_agent_prev_state = cross_prev_state
            cur_agent_prev_action = cross_prev_action
            
            next_agent_strategy = circle_strategy
            next_agent_prev_action = circle_prev_action
            next_agent_prev_state = circle_prev_state
        else:
            cur_agent_strategy = circle_strategy
            cur_agent_prev_state = circle_prev_state
            cur_agent_prev_action = circle_prev_action
            
            next_agent_strategy = cross_strategy
            next_agent_prev_action = cross_prev_action
            next_agent_prev_state = cross_prev_state

        state_hash, empty_cells, cur_turn = env.get_state()
        state_tensor = state_to_tensor(state_hash, env.cols_num)
        
        with torch.no_grad():
            action_idx = cur_agent_strategy.get_action(state_tensor)
            action = env.int_to_action(action_idx) 

        cur_agent_prev_state = state_tensor
        cur_agent_pref_action = action_idx
        
        if env.cur_turn == 1:
            circle_prev_state = state_tensor
            circle_prev_action = action_idx
        else:
            cross_prev_state = state_tensor
            cross_prev_action = action_idx
        

        (next_state, _, cur_turn), reward, is_ended, _ = env.step(action)
        next_state_tensor = state_to_tensor(next_state, env.cols_num)    
        if DEBUG:
            print(
                "cur_turn: ", cur_turn,
                " cur_agent_strategy.name: ", cur_agent_strategy.name,
                " cur_agent_prev_state: ", cur_agent_prev_state,
                " next_state,: ", next_state, 
                " action_idx: ", action_idx,
                " action in env.get_empty_cells(): ", action in env.get_empty_cells(),
                " reward,: ", reward, 
                " is_ende: ", is_ended
        )
        if reward == -10:
            transition = (
                state_tensor,
                action_idx,
                next_state_tensor,
                torch.tensor([NOT_IN_EMPTY_REWARD], dtype=torch.float32)
            )
            cur_agent_strategy.memory.store(transition)
            #print("1:", action_idx)
        else:
            if cur_agent_prev_state is not None and next_agent_prev_state is not None:
                if reward == -1 * cur_turn:
                    transition = (
                        next_agent_prev_state, 
                        next_agent_prev_action, 
                        next_state_tensor, 
                        torch.tensor([1.0], dtype=torch.float32)
                    )
                    next_agent_strategy.memory.store(transition)
                    #print("2:", next_agent_prev_action)

                transition = (
                    cur_agent_prev_state, 
                    cur_agent_prev_action, 
                    next_state_tensor, 
                    torch.tensor([reward * cur_turn], dtype=torch.float32)
                )
                cur_agent_strategy.memory.store(transition)
                #print("3:", cur_agent_prev_action)
            
            state = next_state
        cross_strategy.learn_on_history()
        circle_strategy.learn_on_history()
```

```python
def play_game(env, cross_strategy, circle_strategy):
    env.reset()

    reward = 0
    is_ended = False

    while not env.game_ended:
        state_hash, empty_cells, cur_turn = env.get_state()
        # get current strategy
        if cur_turn == 1:
            cur_agent_strategy = cross_strategy
        else:
            cur_agent_strategy = circle_strategy

        # get current action
        if cur_agent_strategy is not None and state_hash in cur_agent_strategy.Q:
            action_idx = np.argmax(cur_agent_strategy.Q[state_hash])
        else:
            action_idx = np.random.randint(len(empty_cells))
        
        # step
        cur_action = empty_cells[action_idx]
        _, reward, is_ended, _ = env.step(cur_action)
        

        if DEBUG:
            print(cur_action)
            env.print_board()
            print(is_ended)
    
    return reward
```

```python
def play_game(env, cross_strategy, circle_strategy):
    env.reset()

    reward = 0
    is_ended = False
    
    while not env.game_ended:
        state_hash, empty_cells, cur_turn = env.get_state()
        
        # get current strategy
        if cur_turn == 1:
            cur_agent_strategy = cross_strategy
        else:
            cur_agent_strategy = circle_strategy
        
        if cur_agent_strategy is not None:
            action_idx = cur_agent_strategy.get_action(
                state_to_tensor(state_hash, env.cols_num).to(device), True
            )
            print(action_idx)
        else:
            action_idx = np.random.randint(len(empty_cells))

        # step
        cur_action = env.int_to_action(action_idx)
        _, reward, is_ended, _ = env.step(cur_action)
        print(reward, is_ended, action_idx)
        
        if DEBUG:
            print(cur_action)
            env.print_board()
            print(is_ended)
  
    return reward
```

```python
def estimate_reward(env, cross_strategy, circle_strategy, num_episodes=1000):
    acc_reward = 0
    for _ in range(num_episodes):
        acc_reward += play_game(env, cross_strategy, circle_strategy)

    return acc_reward / num_episodes
```

```python
class DQN_3v3(nn.Module):
    def __init__(self, field_size=(3, 3), layer_size=512, hidden_dim=512):
        super().__init__()
        self.model = nn.Sequential(
            nn.Conv2d(in_channels=3, out_channels=layer_size, kernel_size=field_size),
            nn.ReLU(),
            nn.Flatten(),
            nn.Linear(layer_size, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, field_size[0] * field_size[1])
        ) 
        
    def forward(self, x):
        output = self.model(x)
        return output
```

```python
#https://github.com/dronovartem/MADE_rl_dl/blob/main/HW2/hw2.ipynb
from collections import namedtuple

import random
import numpy as np
import tqdm
import torch
import torch.optim as optim
import torch.nn.functional as F
from torch.autograd import Variable

from tictactoe import TicTacToe


class ReplayMemory():
    def __init__(self, capacity):
        self.capacity = capacity
        self.memory = []
        self.position = 0

    def store(self, exptuple):
        if len(self.memory) < self.capacity:
            self.memory.append(None)
        self.memory[self.position] = exptuple
        self.position = (self.position + 1) % self.capacity
       
    def sample(self, batch_size):
        return random.sample(self.memory, batch_size)
    
    def __len__(self):
        return len(self.memory)


class AgentDQN:
    def __init__(
        self,
        model,
        env=None,
        epsilon=0.9,
        gamma=0.8,
        buffer_size=1000000,
        batch_size=512,
        learning_rate=1e-4,
        device='cpu'):

        self.env = env if env is not None else TicTacToe(3, 3, 3)
        self.device = device

        # -1: noughts, +1 crosses
        self.models = {-1: model().to(device), 1: model().to(device)}
        self.memory = {-1: ReplayMemory(buffer_size), 1: ReplayMemory(buffer_size)}
        self.optimizers = {-1: optim.Adam(self.models[-1].parameters(), lr=learning_rate, weight_decay=0.001),
                            1: optim.Adam(self.models[1].parameters(), lr=learning_rate, weight_decay=0.001)} 
        self.epsilon = epsilon
        self.gamma = gamma
        self.batch_size = batch_size


    def __state_to_tensor(self, state):
        size = self.env.n_cols
        state = np.array([int(x) for x in state])
        sx =  np.where(state > 1, 1, 0).reshape(size, size)
        so =  np.where(state < 1, 1, 0).reshape(size, size)
        se = np.where(state == 1, 1, 0).reshape(size, size)
        return torch.Tensor(np.stack([sx, so, se])).reshape(3, size, size)

    def get_action_greedy(self, state, cur_turn):
        return self.models[cur_turn](state.unsqueeze(0)).data.max(1)[1].view(1, 1)

    def get_action(self, state, cur_turn):
        if np.random.random() > self.epsilon:
            return self.get_action_greedy(state, cur_turn)
        else:
            return torch.tensor([[random.randrange(self.env.n_rows * self.env.n_rows)]], dtype=torch.int64)
        
    def __run_episode(self, do_learning=True):
        self.env.reset()
        done = False

        self.prev_states = {-1: None, 1: None}
        self.prev_actions = {}
        state, cur_turn = self.env.getHash(), self.env.curTurn

        while not done:
            state_tensor = self.__state_to_tensor(state)
            with torch.no_grad():
                action_idx = self.get_action(state_tensor.to(self.device), cur_turn).cpu()

            # Сохраняем позицию и совершаемое действие
            self.prev_states[cur_turn] = state_tensor
            self.prev_actions[cur_turn] = action_idx
            action = self.env.action_from_int(action_idx.numpy()[0][0])
            (next_state, _, cur_turn), reward, done, _ = self.env.step(action)
            next_state_tensor = self.__state_to_tensor(next_state)

            if reward == -10:
                transition = (state_tensor, action_idx,
                next_state_tensor, torch.tensor([reward], dtype=torch.float32))
                self.memory[cur_turn].store(transition)
            else:
                if self.prev_states[cur_turn] is not None:
                    if reward == -cur_turn:
                        transition = (self.prev_states[-cur_turn], 
                                      self.prev_actions[-cur_turn], 
                                      next_state_tensor, 
                                      torch.tensor([1.0], dtype=torch.float32)
                                     )
                        self.memory[-cur_turn].store(transition)
                    
                    transition = (self.prev_states[cur_turn], 
                                    self.prev_actions[cur_turn], 
                                    next_state_tensor, 
                                    torch.tensor([reward * cur_turn], dtype=torch.float32)
                                    )
                    self.memory[cur_turn].store(transition)
            
            if do_learning:
                self.update(cur_turn)
            state = next_state


    def update(self, cur_turn):
        # .learn method for 05-policygradient.ipynb
        if len(self.memory[cur_turn]) < self.batch_size:
            return
        
        # берём мини-батч из памяти
        transitions = self.memory[cur_turn].sample(self.batch_size)
        batch_state, batch_action, batch_next_state, batch_reward = zip(*transitions)

        batch_state = Variable(torch.stack(batch_state).to(self.device))
        batch_action = Variable(torch.cat(batch_action).to(self.device))
        batch_reward = Variable(torch.cat(batch_reward).to(self.device))
        batch_next_state = Variable(torch.stack(batch_next_state).to(self.device))
        
        # считаем значения функции Q
        Q = self.models[cur_turn](batch_state)
        Q = Q.gather(1, batch_action).reshape([self.batch_size])
        
        # оцениваем ожидаемые значения после этого действия
        Qmax = self.models[cur_turn](batch_next_state).detach()
        Qmax = Qmax.max(1)[0]
        Qnext = batch_reward + (self.gamma * Qmax)
        # и хотим, чтобы Q было похоже на Qnext -- это и есть суть Q-обучения
        loss = F.smooth_l1_loss(Q, Qnext)
        self.optimizers[cur_turn].zero_grad()
        loss.backward()
        self.optimizers[cur_turn].step()

    
    def _learn_episode(self):
        self.__run_episode(do_learning=True)


    def play_episode(self, random_x=True, random_o=True):
        self.env.reset()
        state, empty_spaces, turn = self.env.getState()
        done = False
        while not done:
            action = None
            if turn == 1:
                if random_x:
                    idx = np.random.randint(len(empty_spaces))
                    action = empty_spaces[idx] 
                else:
                    idx = self.get_action_greedy(self.__state_to_tensor(state).to(self.device), turn)
                    action = self.env.action_from_int(idx)
            elif turn == -1:
                if random_o:
                    idx = np.random.randint(len(empty_spaces))
                    action = empty_spaces[idx] 
                else:
                    idx = self.get_action_greedy(self.__state_to_tensor(state).to(self.device), turn)
                    action = self.env.action_from_int(idx)
            (state, empty_spaces, turn), reward, done, _ = self.env.step(action)
        return reward


    def play(self, n_episodes=1000, random_x=True, random_o=True):
        reward = {0: 0, -1: 0, 1: 0, -10: 0}
        for _ in range(n_episodes):
            reward[self.play_episode(random_x, random_o)] += 1
        for k, _ in reward.items():
            reward[k] /= n_episodes
        return reward


    def learn(self, n_train_episodes=1000, n_test_episodes=1000, verbose=10000, min_eps=0.1):
        Res = namedtuple('result', ['iter', 'x', 'o'])
        rewards_x = []
        rewards_o = []
        n_iter = []

        eps_decay = (self.epsilon - min_eps) / n_train_episodes

        for i in tqdm.tqdm(range(n_train_episodes)):
            self._learn_episode()
            self.epsilon -= eps_decay

            if verbose is not None and (i + 1) % verbose == 0:
                test_reward = self.play(n_episodes=n_test_episodes, random_x=False)
                rewards_x.append(test_reward)
                test_reward = self.play(n_episodes=n_test_episodes, random_o=False)
                rewards_o.append(test_reward)
                n_iter.append(i)

        print("Final X win rate: ",  rewards_x[-1][1])
        print("Final O win rate: ",  rewards_o[-1][-1])
        return Res(n_iter, rewards_x, rewards_o)
    
    d_agent = AgentDQN(model=DQN_4v4, env=TicTacToe(4, 4, 4))
result = d_agent.learn(n_train_episodes=30000, n_test_episodes=1000, verbose=5000, min_eps=0)
plot_results(d_agent, result)
```
