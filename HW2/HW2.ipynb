```python
class TicTacToe(gym.Env):
    def __init__(self, rows_num=ROWS_NUM, cols_num=COLS_NUM, win_len=WIN_LEN):
        self.rows_num = rows_num
        self.cols_num = cols_num
        self.win_len = win_len

        self.reset()

    @staticmethod
    def clone(copy_from):
        new_env = TicTacToe(
            rows_num=copy_from.rows_num, 
            cols_num=copy_from.cols_num,
            win_len=copy_from.win_len
        )

        new_env.board = np.array(copy_from.board, copy=True) 
        new_env.empty_spaces = np.array(copy_from.empty_spaces, copy=True) 
        new_env.board_hash = copy.copy(copy_from.board_hash)
        new_env.game_ended = copy_from.game_ended
        new_env.cur_turn = copy_from.cur_turn

        return new_env

    def get_empty_cells(self, force_update=False):
        if self.empty_spaces is None or force_update:
            res = np.where(self.board == 0)
            self.empty_spaces = np.array(list(zip(res[0], res[1])))
        return self.empty_spaces

    def get_hash(self, force_update=False):
        if self.board_hash is None or force_update:
            self.board_hash = ''.join([
                str(cell + 1) for cell in self.board.reshape(self.rows_num * self.cols_num)
            ])

        return self.board_hash

    def print_board(self):
        print('╭', ('───┬' * self.cols_num)[:-1], '╮', sep='')
        for i in range(0, self.rows_num):
            if i != 0:
                print('├' + ('───┼' * self.cols_num)[:-1] + '┤')
            out = '│ '
            for j in range(0, self.cols_num):
                if self.board[i, j] == 1:
                    out += '✕'
                if self.board[i, j] == -1:
                    out += '◯'
                if self.board[i, j] == 0:
                    out += ' '
                out += ' │ '
            print(out, sep='')
        print('╰' + ('───┴' * self.cols_num)[:-1] + '╯')

    def update(self):
        self.empty_spaces = None
        self.board_hash = None
        self.get_hash()
        self.get_empty_cells()

    def player_turn(self, player, i, j):
        self.board[i, j] = player
        self.update()

    def get_state(self):
        return (self.get_hash(), self.get_empty_cells(), self.cur_turn)

    def int_to_action(self, action_int):
        return (int(action_int / self.cols_num), int(action_int % self.cols_num))

    def action_to_int(self, action):
        return action[0] * self.cols_num + action[1]

    def is_terminal(self):
        cur_marks = np.where(self.board == self.cur_turn)
        cur_position = self.cur_turn

        for i, j in zip(cur_marks[0], cur_marks[1]):
            game_ended = False

            if (
                i <= self.rows_num - self.win_len and
                np.all(self.board[i:i + self.win_len, j] == cur_position)
            ):
                game_ended = True

            if (
                not game_ended and 
                j <= self.cols_num - self.win_len and
                np.all(self.board[i, j:j + self.win_len] == cur_position)
            ):
                game_ended = True

            if (
                not game_ended and
                i <= self.rows_num - self.win_len and j <= self.cols_num - self.win_len and
                np.all(np.array(
                    [self.board[i + k, j + k] == cur_position for k in range(self.win_len)]
                ))
            ):
                game_ended = True

            if (
                not game_ended and
                i <= self.rows_num - self.win_len and j >= self.win_len-1 and
                np.all(np.array([
                    self.board[i + k, j - k] == cur_position for k in range(self.win_len) ]
                ))
            ):
                game_ended = True

            if game_ended:
                self.game_ended = True
                return self.cur_turn

        if len(self.get_empty_cells()) == 0:
            self.game_ended = True
            return 0

        self.game_ended = False
        return None

    def step(self, action):
        if self.board[action[0], action[1]] != 0:
            return self.get_state(), -10, True, {}

        self.player_turn(self.cur_turn, *action)

        reward = self.is_terminal()
        self.cur_turn = -self.cur_turn
        return self.get_state(), 0 if reward is None else reward, reward is not None, {}

    def reset(self):
        self.board = np.zeros((self.rows_num, self.cols_num), dtype=int)
        self.empty_spaces = None

        self.board_hash = None
        self.game_ended = False

        self.cur_turn = 1
```

```python
def learn_episode(env, cross_strategy, circle_strategy, alpha=0.009, epsilon=0.1, gamma=1.0):

    env.reset()
    
    cross_prev_action = None
    cross_prev_board_state = None

    circle_prev_action = None
    circle_prev_board_state = None

    while not env.game_ended:
        
        # get current strategy
        if env.cur_turn == 1:
            cur_agent_strategy = cross_strategy
            next_agent_strategy = circle_strategy

            next_agent_prev_action = circle_prev_action
            next_agent_prev_board_state = circle_prev_board_state
        else:
            cur_agent_strategy = circle_strategy
            next_agent_strategy = cross_strategy

            next_agent_prev_action = cross_prev_action
            next_agent_prev_board_state = cross_prev_board_state

        state_hash, empty_cells, cur_turn = env.get_state()
        
        # get current action
        if cur_agent_strategy is None :
            idx = np.random.randint(len(empty_cells))
        else: 
            if state_hash in cur_agent_strategy.Q and np.random.random() > epsilon: 
                # action by strategy
                idx = np.argmax(cur_agent_strategy.Q[state_hash])
            else:
                # action by exploration
                idx = np.random.randint(len(empty_cells))
                if state_hash not in cur_agent_strategy.Q:
                    # explore new state
                    cur_agent_strategy.Q[state_hash] = np.zeros(len(empty_cells))
        if DEBUG:       
            print("cur_agent_strategy: {}".format(cur_agent_strategy.name))
            print("idx: {}".format(idx))
            print("empty_cells: {}".format(empty_cells))
            print("empty_cells[idx]: {}".format(idx))

        cur_action = empty_cells[idx]
        (next_state_hash, _, _), reward, is_ended, _ = env.step(cur_action)

        # train next player straregy
        if next_agent_strategy is not None and next_agent_prev_board_state is not None: 

            gamma_term = 0
            if next_state_hash in next_agent_strategy.Q:
                gamma_term = gamma * np.max(next_agent_strategy.Q[next_state_hash])
            
            if DEBUG:
                print(
                    "next_agent_prev_action: {}".format(
                        next_agent_prev_action
                    )
                )
                print(
                    "next_agent_prev_board_state: {}".format(
                        next_agent_prev_board_state
                    )
                )
                print(
                    "next_agent_strategy.Q[next_agent_prev_board_state]: {}".format(
                        next_agent_strategy.Q[next_agent_prev_board_state]
                    )
                )
                print(
                    "next_agent_strategy.Q[next_agent_prev_board_state][next_agent_prev_action]: {}".format(
                        next_agent_strategy.Q[next_agent_prev_board_state][next_agent_prev_action]
                    )
                )
            prev_q_value = next_agent_strategy.Q[next_agent_prev_board_state][next_agent_prev_action]

            #cur_reward - reward of current agent: cur_turn * reward 
            # if it's crosses turn: 1 * reward, else - (-1) * (-reward)
            cur_reward = cur_turn * reward
            new_q = (cur_reward + gamma_term)
            err = (new_q - prev_q_value)
            next_agent_strategy.Q[next_agent_prev_board_state][next_agent_prev_action] = \
                prev_q_value + alpha * err

        # this is good state for curent agent, because it has won
        if cur_agent_strategy is not None and abs(reward) == 1:
              cur_agent_strategy.Q[state_hash][idx] = abs(reward)

        if env.cur_turn == 1:
            circle_prev_board_state = state_hash
            circle_prev_action = idx
        else:
            cross_prev_board_state = state_hash
            cross_prev_action = idx
```

```python
def play_game(env, cross_strategy, circle_strategy):
    env.reset()

    reward = 0
    is_ended = False

    while not is_ended:
        state_hash, empty_cells, cur_turn = env.get_state()
        # get current strategy
        if cur_turn == 1:
            cur_agent_strategy = cross_strategy
        else:
            cur_agent_strategy = circle_strategy

        # get current action
        if cur_agent_strategy is not None and state_hash in cur_agent_strategy.Q:
            action_idx = np.argmax(cur_agent_strategy.Q[state_hash])
        else:
            action_idx = np.random.randint(len(empty_cells))
        
        # step
        cur_action = empty_cells[action_idx]
        _, reward, is_ended, _ = env.step(cur_action)
        

        if DEBUG:
            print(cur_action)
            env.print_board()
            print(is_ended)
    
    return reward
```

```python
def estimate_reward(env, cross_strategy, circle_strategy, num_episodes=1000):
    acc_reward = 0
    for _ in range(num_episodes):
        acc_reward += play_game(env, cross_strategy, circle_strategy)

    return acc_reward / num_episodes
```

```python
def plot_stats(crosse_scores, circle_scores, test_episodes):
    display.clear_output(wait=True)
    
    fig = plt.figure(figsize=(18, 8))
    plt.xlabel('Number of episodes')
    plt.ylabel('Average reward')
    plt.plot(range(0, len(crosse_scores) * test_episodes, test_episodes), crosse_scores, label='Crosses')
    plt.plot(range(0, len(circle_scores) * test_episodes, test_episodes), circle_scores, label='Circles')
    plt.legend(loc="upper left")
    plt.grid(True)
    plt.show()
```

```python
train_episodes_num = 100000
test_episodes_num = 1000

env = TicTacToe(3, 3, 3)
cross_strategy = Strategy("cross_strategy")
circle_strategy = Strategy("circle_strategy")

crosse_scores = []
circle_scores = []
fig = plt.figure(figsize=(18, 8))
plt.show()
for i in tqdm.tqdm(range(1, train_episodes_num + 1)):
    learn_episode(env, None, circle_strategy, alpha=0.001, epsilon=0.4, gamma=1)
    learn_episode(env, cross_strategy, None, alpha=0.001, epsilon=0.4, gamma=1)
    learn_episode(env, cross_strategy, circle_strategy, alpha=0.001, epsilon=0.4, gamma=1)
    if i % test_episodes_num == 0:
        crosse_scores.append(estimate_reward(env, cross_strategy, None, num_episodes=500))
        circle_scores.append(-1 * estimate_reward(env, None, circle_strategy, num_episodes=500))
        plot_stats(crosse_scores, circle_scores, test_episodes_num)
```

```python
EXPLORATION_PARAM = 2 # 'c'

class Node:
    def __init__(self, action=None, cur_turn=None, parent=None, empty_cells=None):
        self.turn = cur_turn
        self.action = action
        self.parent_node = parent

        self.wins = 0
        self.visits = 0

        self.child_nodes = {}
        self.unchecked_actions = [tuple(cell) for cell in empty_cells]

    def mcts_selection(self, node):
        return (
            node.wins / node.visits + 
            EXPLORATION_PARAM * np.sqrt(2 * np.log(self.visits) / node.visits)
        )

    def select_child(self):
        return max(
            self.child_nodes.values(),
            key=self.mcts_selection
        )

    def add_child(self, action, empty_cells, cur_turn):
        node = Node(
            action=action,
            cur_turn=cur_turn,
            parent=self,
            empty_cells=empty_cells
        )

        self.unchecked_actions.remove(action)
        self.child_nodes[action] = node

        return node

    def add_win(self, reward):
        self.visits += 1
        self.wins += reward * self.turn

def rollout(env, policy_crosses, policy_circles):
    clone_env = TicTacToe.clone(env)
    is_ended = False
    while not is_ended:
        state = env.get_hash()
        if env.cur_turn == 1 and state in policy_crosses:
            action = policy_crosses[state]
        elif env.cur_turn == 0 and state in policy_circles:
            action = policy_circles[state]
        else:
            action = get_random_action(clone_env)
        _, reward, is_ended, _ = clone_env.step(action)
    return reward

def get_random_action(env):
    empty_cells = env.get_empty_cells()
    action_idx = np.random.randint(len(empty_cells))
    return tuple(empty_cells[action_idx])

def play_game(env, policy, policy_turn):
    env.reset()
    
    is_ended = False
    policy_exists = True

    state = env.get_hash()
    cur_turn = env.cur_turn
    
    while not is_ended:
        if env.cur_turn == policy_turn:
            if len(policy.child_nodes) > 0 and policy_exists:
                policy = max(policy.child_nodes.values(), key=lambda x: x.wins / x.visits)
                _, reward, is_ended, _ = env.step(policy.action)
            else:
                action = get_random_action(env)
                _, reward, is_ended, _ = env.step(action)
        else:
            action = get_random_action(env)
            _, reward, is_ended, _ = env.step(action)
            if action in policy.child_nodes:
                policy = policy.child_nodes[action]
            else:
                policy_exists = False
        if DEBUG:
            print(cur_action)
            env.print_board()
            print(is_ended)
            
    return reward

def estimate_reward(env, policy_turn, policy, num_episodes=1000):
    rewards = {1: 0, 0: 0, -1: 0}
    for _ in range(num_episodes):
        reward = play_game(env, policy, policy_turn)
        rewards[reward] = 1 + rewards[reward]
 
    return rewards

def train_episode(env, rootnode):
    reward = 0
    is_ended = False
    env.reset()
    node = rootnode

    while len(node.unchecked_actions) == 0 and len(node.child_nodes) > 0:
        node = node.select_child()
        _, reward, is_ended, _ = env.step(node.action)

    if len(node.unchecked_actions) > 0:
        action = random.choice(node.unchecked_actions)
        cur_turn = env.cur_turn
        asd, reward, is_ended, _ = env.step(action)
        node = node.add_child(
            action=action,
            empty_cells=env.get_empty_cells() if not env.game_ended else [],
            cur_turn=cur_turn
        )

    # rollout
    #if not is_ended:
    #    reward = rollout(env, policy_crosses, policy_circles)

    while node is not None:
        node.add_win(reward)
        node = node.parent_node
```
