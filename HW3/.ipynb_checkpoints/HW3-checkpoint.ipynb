```python
class FacadesDataset(Dataset):
    def __init__(self, path, required_size=None):
        self.base_path = path
        self.files_list = os.listdir(path)
        self.required_size = required_size

    def __len__(self):
        return len(self.files_list)

    def __getitem__(self, img_idx):
        filename = self.files_list[img_idx]

        image = Image.open(os.path.join(self.base_path, filename))
        image = transforms.functional.to_tensor(image)

        image_width = image.shape[2]

        condition = image[:, :, image_width // 2 :]
        img_data = image[:, :, : image_width // 2]

        if self.required_size:
            condition = nn.functional.interpolate(condition, size=self.required_size)
            img_data = nn.functional.interpolate(img_data, size=self.required_size)

        return img_data, condition
```

```python
def conv_output_shape(h_w, kernel_size=1, strides=1, padding=0, dilation=1):
    
    if type(h_w) is not tuple:
        h_w = (h_w, h_w)
    
    if type(kernel_size) is not tuple:
        kernel_size = (kernel_size, kernel_size)
    
    if type(strides) is not tuple:
        strides = (strides, strides)
    
    if type(padding) is not tuple:
        padding = (padding, padding)
    
    h = (h_w[0] + (2 * padding[0]) - (dilation * (kernel_size[0] - 1)) - 1) // strides[0] + 1
    w = (h_w[1] + (2 * padding[1]) - (dilation * (kernel_size[1] - 1)) - 1) // strides[1] + 1
    
    return h, w


def init_weights(model):
    if isinstance(model, (nn.Conv2d, nn.ConvTranspose2d, nn.BatchNorm2d)):
        model.weight.data.normal_(0., 0.02)
        model.bias.data.zero_()


class DownConvBlock(nn.Module):
    def __init__(
        self, 
        in_channels,
        out_channels,
        kernel_size=4,
        strides=2,
        padding=1, 
        batchnorm=True
    ):
        super().__init__()

        self.in_channels = in_channels
        self.out_channels = out_channels
        self.kernel_size = kernel_size
        self.strides = strides
        self.padding = padding
        
        self.conv = nn.Conv2d(in_channels, out_channels, kernel_size, strides, padding)

        if batchnorm:
            self.batchnorm = nn.BatchNorm2d(out_channels)
        else:
            self.batchnorm = None

    def forward(self, input):
        output = self.conv(input)
        if self.batchnorm:
            output = self.batchnorm(output)

        return output
    
    def __str__(self):
        return "in_channels={}, out_channels={}, kernel_size={}, strides={}, padding={} -> {} * {}".format(
            self.in_channels,
            self.out_channels,
            self.kernel_size,
            self.strides,
            self.padding,
            *conv_output_shape((256, 256), self.kernel_size, self.strides, self.padding)
         )


class UpConvBlock(nn.Module):

    def __init__(self, in_channels, out_channels, 
                 kernel_size=4, strides=2, padding=1, 
                 batchnorm=True, dropout=False):
        super().__init__()
                   
        self.in_channels = in_channels
        self.out_channels = out_channels
        self.kernel_size = kernel_size
        self.strides = strides
        self.padding = padding

        self.deconv = nn.ConvTranspose2d(in_channels, out_channels, kernel_size, strides, padding)

        if batchnorm:
            self.batchnorm = nn.BatchNorm2d(out_channels)
        else:
            self.batchnorm = None

        if dropout:
            self.dropout = nn.Dropout(0.5)
        else:
            self.dropout = None

    def forward(self, x):
        x = self.deconv(x)
        if self.batchnorm:
            x = self.batchnorm(x)

        if self.dropout:
            x = self.dropout(x)
        return x
    
    def __str__(self):
        return "in_channels={}, out_channels={}, kernel_size={}, strides={}, padding={} -> {} * {}".format(
            self.in_channels,
            self.out_channels,
            self.kernel_size,
            self.strides,
            self.padding,
            *conv_output_shape((256, 256), self.kernel_size, self.strides, self.padding)
        )
```

```python
class Generator(nn.Module):

    def __init__(self, in_channels=3, out_channels=3):
        super().__init__()
        
        self.down_relu = nn.LeakyReLU(0.2, True)
        self.up_relu = nn.ReLU(True)
        self.tanh = nn.Tanh()

        self.encoder_layers = nn.ModuleList([
            DownConvBlock(in_channels, 64, batchnorm=False),  # x 128
            DownConvBlock(64, 128),                           # x 64
            DownConvBlock(128, 256),                          # x 32
            DownConvBlock(256, 512),                          # x 16
            DownConvBlock(512, 512),                          # x 8
            DownConvBlock(512, 512),                          # x 4
            DownConvBlock(512, 512),                          # x 2
            DownConvBlock(512, 512, batchnorm=False),         # x 1
        ])

        self.decoder_layers = nn.ModuleList([
            UpConvBlock(512, 512, dropout=True),  # bs x 512 x 2 x 2
            UpConvBlock(1024, 512, dropout=True),  # bs x 512 x 4 x 4
            UpConvBlock(1024, 512, dropout=True),  # bs x 512 x 8 x 8
            UpConvBlock(1024, 512),  # bs x 512 x 16 x 16
            UpConvBlock(1024, 256),  # bs x 256 x 32 x 32
            UpConvBlock(512, 128),  # bs x 128 x 64 x 64
            UpConvBlock(256, 64),  # bs x 64 x 128 x 128
            UpConvBlock(128, out_channels, batchnorm=False)
        ])

    def forward(self, input):
        encoder_output = self.encoder_layers[0](input)
        encoder_output_list = [encoder_output]
        for encoder_layer in self.encoder_layers[1:]:
            encoder_output = encoder_layer(self.down_relu(encoder_output))
            encoder_output_list.append(encoder_output)
            
        encoder_skips = encoder_output_list[::-1]
        decoder_output = self.decoder_layers[0](self.up_relu(encoder_skips[0]))
        for i in range(1, len(self.decoder_layers)):
            decoder_output_skip = torch.cat([decoder_output, encoder_skips[i]], 1)
            decoder_output = self.decoder_layers[i](self.up_relu(decoder_output_skip))
            
        return self.tanh(decoder_output)
```

```python
class Discriminator(nn.Module):

    def __init__(self, in_out_channels=3+3):
        super().__init__()
        
        self.model = nn.Sequential(
            DownConvBlock(in_out_channels, 64, batchnorm=False),
            nn.LeakyReLU(0.2, True),
            DownConvBlock(64, 128),
            nn.LeakyReLU(0.2, True),
            DownConvBlock(128, 256),
            nn.LeakyReLU(0.2, True),
            DownConvBlock(256, 512),
            nn.LeakyReLU(0.2, True),
            nn.Conv2d(512, 1, kernel_size=4)
        )

    def forward(self, x, y):
        x = torch.cat([x, y], axis=1)
        x = self.model(x)
        return x
```

```python
class Pix2Pix(nn.Module):
    def __init__(self, in_channels, out_channels, learning_rate=0.0002, lambda_recon=200, device = device):
        super().__init__()

        self.device = device

        self.generator = Generator(in_channels, out_channels)
        self.generator.apply(init_weights)
        self.generator.to(self.device)
    
        self.discriminator = Discriminator(in_channels + out_channels)
        self.discriminator.apply(init_weights)
        self.discriminator.to(self.device)
        
        self.adversarial_criterion = nn.BCEWithLogitsLoss()
        self.reconstruct_criterion = nn.L1Loss()
        
        self.generator_optimizer = torch.optim.Adam(self.generator.parameters(), lr=learning_rate)
        self.discriminator_optimizer = torch.optim.Adam(self.discriminator.parameters(), lr=learning_rate)
        
        self.lambda_recon = lambda_recon
        
    def apply(self, batch, train=False):
        if train:
            self.generator.train()
            self.discriminator.train()
        else:
            self.generator.eval()
            self.discriminator.eval()

        original_images = batch[0].to(self.device)
        condition_images = batch[1].to(self.device)
        generated_images = self.generator(condition_images)

        discriminator_logits = self.discriminator(generated_images, condition_images)
        adversarial_loss = self.adversarial_criterion(discriminator_logits, torch.ones_like(discriminator_logits))
        reconstruct_loss = self.reconstruct_criterion(generated_images, original_images)
        generator_loss = adversarial_loss + self.lambda_recon * reconstruct_loss

        if train:
            self.generator_optimizer.zero_grad()
            generator_loss.backward()
            self.generator_optimizer.step()

        generated_images = self.generator(condition_images).detach()
        generator_logits = self.discriminator(generated_images, condition_images)
        original_logits = self.discriminator(original_images, condition_images)

        generator_loss = self.adversarial_criterion(generator_logits, torch.zeros_like(generator_logits))
        original_loss = self.adversarial_criterion(original_logits, torch.ones_like(original_logits))
        discriminator_loss = (original_loss + generator_loss) / 2
        if train:
            self.discriminator_optimizer.zero_grad()
            discriminator_loss.backward()
            self.discriminator_optimizer.step()

        return generator_loss.cpu().item(), discriminator_loss.cpu().item()
    
    def train(self, batch):
        return self.apply(batch, train=True)

    def evaluate(self, batch):
        return self.apply(batch, train=False)
    
    def generate(self, batch):
        self.generator.eval()
        self.discriminator.eval()
        
        original_images, condition_images = batch
        original_images = original_images.to(self.device)
        condition_images = condition_images.to(self.device)

        with torch.no_grad():
            generated_images = self.generator(condition_images)

        return original_images, condition_images, generated_images
```

```python
def plot_stats(
    train_gen_loss, train_disc_loss, 
    test_gen_loss, test_disc_loss, 
    test_episodes,
    original_img,
    condition_img,
    generated_img
):
    display.clear_output(wait=True)   
    
    def format_axes(fig):
        for i, ax in enumerate(fig.axes):
            ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
            ax.tick_params(labelbottom=False, labelleft=False)

    fig = plt.figure(constrained_layout=True, figsize=(18, 10))

    gs = GridSpec(5, 7, figure=fig)
    desc_ax = fig.add_subplot(gs[0:3, 0:3])
    gen_ax =  fig.add_subplot(gs[0:3, 3:6])

    original_ax = fig.add_subplot(gs[3:5, 0:2])
    condition_ax = fig.add_subplot(gs[3:5, 2:4])
    generated_ax = fig.add_subplot(gs[3:5, 4:6])

    #format_axes(fig)
    
    gen_ax.plot(range(0, len(train_gen_loss)), train_gen_loss, label="train_gen_loss")
    gen_ax.plot(range(0, len(test_gen_loss)), test_gen_loss, label="test_gen_loss")
    gen_ax.set_xlabel("Number of epochs")
    gen_ax.set_ylabel("Loss")
    gen_ax.legend(loc="upper left")
    gen_ax.grid(True)
    
    desc_ax.plot(range(0, len(train_disc_loss)), train_disc_loss, label="train_disc_loss")
    desc_ax.plot(range(0, len(test_disc_loss)), test_disc_loss, label="test_disc_loss")
    desc_ax.set_xlabel("Number of epochs")
    desc_ax.set_ylabel("Loss")
    desc_ax.legend(loc="upper left")
    desc_ax.grid(True)
    
    original_img = original_img.detach().cpu().permute(1, 2, 0)
    condition_img = condition_img.detach().cpu().permute(1, 2, 0)
    generated_img = generated_img.detach().cpu().permute(1, 2, 0)
    
    original_ax.imshow(torch.clip(original_img, 0, 1))
    original_ax.set_xlabel("Original Image")
    original_ax.tick_params(labelbottom=False, labelleft=False)
    
    condition_ax.imshow(torch.clip(condition_img, 0, 1))
    condition_ax.set_xlabel("Condidtion image")
    condition_ax.tick_params(labelbottom=False, labelleft=False)
    
    generated_ax.imshow(torch.clip(generated_img, 0, 1))
    generated_ax.set_xlabel("Generated image")
    generated_ax.tick_params(labelbottom=False, labelleft=False)
    
    plt.show()
```

```python
img_size = 256
batch_size = 5

train = FacadesDataset('./facades/train', required_size=img_size)
train_dataloader = DataLoader(train, batch_size=batch_size, shuffle=True)

test = FacadesDataset('./facades/test', required_size=img_size)
test_dataloader = DataLoader(test, batch_size=batch_size, shuffle=True)
```

```python
epochs_num = 200
test_episodes = 10

lambda_recon = 200
lr = 3e-4

train_gen_loss = []
train_disc_loss = []
test_gen_loss = []
test_disc_loss = []


pix2pix = Pix2Pix(3, 3, learning_rate=3e-4, lambda_recon=lambda_recon)

for i in tqdm.tqdm(range(epochs_num)):
    cur_train_gen_loss = []
    cur_train_disc_loss = []
    cur_test_gen_loss = []
    cur_test_disc_loss = []
    
    for batch in train_dataloader:
        gen_loss, disc_loss = pix2pix.train(batch)
        cur_train_gen_loss.append(gen_loss)
        cur_train_disc_loss.append(disc_loss)
    
    for batch in test_dataloader:
        gen_loss, disc_loss = pix2pix.evaluate(batch)
        cur_test_gen_loss.append(gen_loss)
        cur_test_disc_loss.append(disc_loss)
    
    if i % test_episodes == 0:
        original_img, condition_img, generated_img = pix2pix.generate(batch)
        
        train_gen_loss.append(np.mean(cur_train_gen_loss))
        train_disc_loss.append(np.mean(cur_train_disc_loss))
        test_gen_loss.append(np.mean(cur_test_gen_loss))
        test_disc_loss.append(np.mean(cur_test_disc_loss))
        
        plot_stats(
            train_gen_loss, train_disc_loss, 
            test_gen_loss, test_disc_loss, 
            test_episodes,
            original_img[-1],
            condition_img[-1],
            generated_img[-1]
        )
```

# Часть вторая: добавим циклы
Вторая часть -- реализация CycleGAN. Предлагается на том же датасете реализовать базовую конструкцию CycleGAN для unpaired style transfer, то есть переноса стилей, каждый из которых задан неразмеченным датасетом (подробнее см. лекцию и собственно Zhu et al., 2017):

<img src="scheme_2.png" width="600"/>

```python
class CycleGaN(nn.Module):

    def __init__(self, in_channels, out_channels, learning_rate=0.0002, lambda_recon=200, device=device):
        super().__init__()
        self.device = device

        self.generator_a = Generator(in_channels, out_channels) # x
        self.generator_a.apply(init_weights)
        self.generator_a.to(self.device)
        
        self.generator_b = Generator(out_channels, in_channels) # y
        self.generator_b.apply(init_weights)
        self.generator_b.to(self.device)

        self.discriminator_a = Discriminator(in_channels + out_channels)
        self.discriminator_a.apply(init_weights)
        self.discriminator_a.to(self.device)

        self.discriminator_b = Discriminator(in_channels + out_channels)
        self.discriminator_b.apply(init_weights)
        self.discriminator_b.to(self.device)

        self.adv_criterion = nn.BCEWithLogitsLoss()
        self.recon_criterion = nn.L1Loss()
        
        self.generator_optimizer = torch.optim.Adam(
            chain(self.generator_a.parameters(), self.generator_b.parameters()), lr=learning_rate
        )
        self.discriminator_optimizer = torch.optim.Adam(
            chain(self.discriminator_a.parameters(), self.discriminator_b.parameters()), lr=learning_rate
        )
        
        self.lambda_recon = lambda_recon

    def apply(self, batch, train=False):
        if train:
            self.generator_a.train()
            self.generator_b.train()
            self.discriminator_a.eval()
            self.discriminator_b.eval()
        else:
            self.generator_a.eval()
            self.generator_b.eval()
            self.discriminator_a.eval()
            self.discriminator_b.eval()

        original_img_a = batch[1].to(self.device)
        original_img_b = batch[0].to(self.device)

        generated_img_b = self.generator_a(original_img_a)
        generated_img_a = self.generator_b(original_img_b)

        reconv_img_a = self.generator_b(generated_img_b)
        reconv_img_b = self.generator_a(generated_img_a)
        
        loss_idt_a = self.recon_criterion(generated_img_b, original_img_b) * self.lambda_recon
        loss_idt_b = self.recon_criterion(generated_img_a, original_img_a) * self.lambda_recon
        
        discr_a_logits = self.discriminator_a(generated_img_b, original_img_a).detach()
        discr_b_logits = self.discriminator_b(generated_img_a, original_img_b).detach()

        loss_gen_a = self.adv_criterion(discr_a_logits, torch.ones_like(discr_a_logits))
        loss_gen_b = self.adv_criterion(discr_b_logits, torch.ones_like(discr_b_logits))
        
        loss_cyc_a = self.recon_criterion(reconv_img_a, original_img_a) * self.lambda_recon
        loss_cyc_b = self.recon_criterion(reconv_img_b, original_img_b) * self.lambda_recon
        
        loss_full = loss_gen_a + loss_gen_b + loss_cyc_a + loss_cyc_b + loss_idt_a + loss_idt_b

        if train:
            self.generator_optimizer.zero_grad()
            loss_full.backward()
            self.generator_optimizer.step()

            self.generator_a.eval()
            self.generator_b.eval()
            self.discriminator_a.train()
            self.discriminator_b.train()
        
        generated_img_b = self.generator_a(original_img_a).detach()
        generated_img_a = self.generator_b(original_img_b).detach()
        
        generated_logits_a = self.discriminator_a(generated_img_b, original_img_a)
        original_logits_a = self.discriminator_a(original_img_b, original_img_a)
        generated_loss_a = self.adv_criterion(generated_logits_a, torch.zeros_like(generated_logits_a))
        original_loss_a = self.adv_criterion(original_logits_a, torch.ones_like(original_logits_a))
        discriminator_a_loss = (original_loss_a + generated_loss_a) / 2
        
        generated_logits_b = self.discriminator_b(generated_img_a, original_img_b)
        original_logits_b = self.discriminator_b(original_img_a, original_img_b)
        generated_loss_b = self.adv_criterion(generated_logits_b, torch.zeros_like(generated_logits_b))
        original_loss_b = self.adv_criterion(original_logits_b, torch.ones_like(original_logits_b))
        discriminator_b_loss = (original_loss_b + generated_loss_b) / 2

        if train:
            self.discriminator_optimizer.zero_grad()
            discriminator_a_loss.backward()
            discriminator_b_loss.backward()
            self.discriminator_optimizer.step()

        return loss_full.cpu().item(), discriminator_a_loss.cpu().item(), discriminator_b_loss.cpu().item()
    
    def train(self, batch):
        return self.apply(batch, train=True)

    def evaluate(self, batch):
        return self.apply(batch, train=False)

    def generate(self, batch):
        self.generator_a.eval()
        self.generator_b.eval()
        self.discriminator_a.eval()
        self.discriminator_b.eval()
        
        original_images, condition_images = batch
        original_images = original_images.to(self.device)
        condition_images = condition_images.to(self.device)
        
        with torch.no_grad():
            generated_images = self.generator_a(condition_images)
        
        return original_images, condition_images, generated_images
```

```python
def plot_stats(
    train_gen_loss, train_disc_a_loss, train_disc_b_loss,
    test_gen_loss, test_disc_a_loss, test_disc_b_loss, 
    test_episodes,
    original_img,
    condition_img,
    generated_img
):
    display.clear_output(wait=True)   
    
    def format_axes(fig):
        for i, ax in enumerate(fig.axes):
            ax.text(0.5, 0.5, "ax%d" % (i+1), va="center", ha="center")
            ax.tick_params(labelbottom=False, labelleft=False)

    fig = plt.figure(constrained_layout=True, figsize=(18, 10))

    gs = GridSpec(5, 7, figure=fig)
    desc_a_ax = fig.add_subplot(gs[0:3, 0:2])
    desc_b_ax = fig.add_subplot(gs[0:3, 2:4])
    gen_ax =  fig.add_subplot(gs[0:3, 4:6])

    original_ax = fig.add_subplot(gs[3:5, 0:2])
    condition_ax = fig.add_subplot(gs[3:5, 2:4])
    generated_ax = fig.add_subplot(gs[3:5, 4:6])

    #format_axes(fig)
    
    gen_ax.plot(range(0, len(train_gen_loss)), train_gen_loss, label="train_gen_loss")
    gen_ax.plot(range(0, len(test_gen_loss)), test_gen_loss, label="test_gen_loss")
    gen_ax.set_xlabel("Number of epochs")
    gen_ax.set_ylabel("Loss")
    gen_ax.legend(loc="upper left")
    gen_ax.grid(True)
    
    desc_a_ax.plot(range(0, len(train_disc_a_loss)), train_disc_a_loss, label="train_disc_a_loss")
    desc_a_ax.plot(range(0, len(test_disc_a_loss)), test_disc_a_loss, label="test_disc_a_loss")
    desc_a_ax.set_xlabel("Number of epochs")
    desc_a_ax.set_ylabel("Loss")
    desc_a_ax.legend(loc="upper left")
    desc_a_ax.grid(True)
    
    desc_b_ax.plot(range(0, len(train_disc_b_loss)), train_disc_b_loss, label="train_disc_b_loss")
    desc_b_ax.plot(range(0, len(test_disc_b_loss)), test_disc_b_loss, label="test_disc_b_loss")
    desc_b_ax.set_xlabel("Number of epochs")
    desc_b_ax.set_ylabel("Loss")
    desc_b_ax.legend(loc="upper left")
    desc_b_ax.grid(True)
    
    original_img = original_img.detach().cpu().permute(1, 2, 0)
    condition_img = condition_img.detach().cpu().permute(1, 2, 0)
    generated_img = generated_img.detach().cpu().permute(1, 2, 0)
    
    original_ax.imshow(torch.clip(original_img, 0, 1))
    original_ax.set_xlabel("Original Image")
    original_ax.tick_params(labelbottom=False, labelleft=False)
    
    condition_ax.imshow(torch.clip(condition_img, 0, 1))
    condition_ax.set_xlabel("Condidtion image")
    condition_ax.tick_params(labelbottom=False, labelleft=False)
    
    generated_ax.imshow(torch.clip(generated_img, 0, 1))
    generated_ax.set_xlabel("Generated image")
    generated_ax.tick_params(labelbottom=False, labelleft=False)
    
    plt.show()
```

```python
img_size = 256
batch_size = 10

train = FacadesDataset('./facades/train', required_size=img_size)
train_dataloader = DataLoader(train, batch_size=batch_size, shuffle=True)

test = FacadesDataset('./facades/test', required_size=img_size)
test_dataloader = DataLoader(test, batch_size=batch_size, shuffle=True)
```

```python
epochs_num = 500
test_episodes = 10

lambda_recon = 200
lr = 3e-4

train_gen_loss = []
train_disc_a_loss = []
train_disc_b_loss = []

test_gen_loss = []
test_disc_a_loss = []
test_disc_b_loss = []


cyclePix2pix = CycleGaN(3, 3, learning_rate=3e-4, lambda_recon=lambda_recon, device=device)

for i in tqdm.tqdm(range(epochs_num)):
    cur_train_gen_loss = []
    cur_train_disc_a_loss = []
    cur_train_disc_b_loss = []
    
    cur_test_gen_loss = []
    cur_test_disc_a_loss = []
    cur_test_disc_b_loss = []
    
    for batch in train_dataloader:
        gen_loss, disc_a_loss, disc_b_loss = cyclePix2pix.train(batch)
        cur_train_gen_loss.append(gen_loss)
        cur_train_disc_a_loss.append(disc_a_loss)
        cur_train_disc_b_loss.append(disc_b_loss)
    
    for batch in test_dataloader:
        gen_loss, disc_a_loss, disc_b_loss = cyclePix2pix.evaluate(batch)
        cur_test_gen_loss.append(gen_loss)
        cur_test_disc_a_loss.append(disc_a_loss)
        cur_test_disc_b_loss.append(disc_b_loss)
    
    if i % test_episodes == 0:
        original_img, condition_img, generated_img = cyclePix2pix.generate(batch)
        
        train_gen_loss.append(np.mean(cur_train_gen_loss))
        test_gen_loss.append(np.mean(cur_test_gen_loss))
        
        train_disc_a_loss.append(np.mean(cur_train_disc_a_loss))
        test_disc_a_loss.append(np.mean(cur_test_disc_a_loss))
        
        train_disc_b_loss.append(np.mean(cur_train_disc_b_loss))
        test_disc_b_loss.append(np.mean(cur_test_disc_b_loss))
        
        plot_stats(
            train_gen_loss, train_disc_a_loss, train_disc_b_loss,
            test_gen_loss, test_disc_a_loss, test_disc_b_loss, 
            test_episodes,
            original_img[-1],
            condition_img[-1],
            generated_img[-1]
        )
```
